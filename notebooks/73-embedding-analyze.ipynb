# imports

```python
from gensim.models import Word2Vec
```

# constants

```python
W2V_MODEL = "../data/word2vec.model" # NOT STORED IN REPOSITORY; GENERATED BY NOTEBOOK
```

# analyze

```python
w2v_model = Word2Vec.load(W2V_MODEL)
print(f"Number of words in vocabulary: {len(w2v_model.wv)}")
```

```python
print(w2v_model.wv.most_similar('clean', topn=10))
```

```python
print(w2v_model.wv.similarity('clean', 'tidi'))
```
